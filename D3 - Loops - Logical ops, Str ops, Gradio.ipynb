## Logical Operations
- Boolean logic system - electronics, mathematics, number systems
- exists on two symbol number systems
- Binary Number System
  - 0, 1
  - False, True
  - OFF, ON
  - 0V, 5V

- Logical operations came to existence for Bool values
- might work for other data types, but are most important with Booleans

- Operators:

OR

AND

NOT

```python
print("TRUTH TABLE FOR ---- OR OPERATION")
print("False or False =", False or False)
print("False or True =", False or True)
print("True or False =", True or False)
print("True or True =", True or True)

print("-" * 100)

print("TRUTH TABLE FOR ---- AND OPERATION")
print("False and False =", False and False)
print("False and True =", False and True)
print("True and False =", True and False)
print("True and True =", True and True)
```

## String Operations
- String - is a data type in python
- if you store a variable:
- Are **Primitive data types**: can be of integer, float, string, boolean, None
    - the most basic data types
- **Sequence data types**: - That have index
    - are in which data is stored in a sequence and has indices
    - String
    - List
    - Tuple
- **Other Data types**:
    - Set
    - Dictionary

-
- String operations are there to help us deal with user input

```python
# what is an index

name = "Siti Alya Nurjannah"
name[6]
```

```python
# take an input from the user
# let's say we have to do a Google search
# or we have to prompt an LLM - ChatGPT, or Gemini etc

prompt = input("Query:")

data = prompt

# string operations help us clean the users inputs
```

```python
data
```

```python
print(data.lstrip()) # removes the leading spaces - LEFT
print(data.rstrip()) # removes trailing spaces - RIGHT
print(data.strip()) # removes useless spaces on left and right

print("-" * 100)

print(data.upper())
print(data.lower())

print("-" * 100)

# search for something
print(data.find("tr")) # -1 if not found
print(data.find("Tr")) # returns the index of the word
print(data.find("p"))

print("-" * 100)

print(data.count("R"))
print(data.count("r"))
print(data.count(" "))
print(data.count("  "))
```
