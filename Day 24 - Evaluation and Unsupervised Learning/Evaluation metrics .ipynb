Precision => used when False Positives are important and cannot be missed (avoid False Alarms)
eg. bad customer experience - bringing non smoker ashtray because model said smoker

Recall => used when False Negatives are important and cannot be missed (avoid Missed Cases)
eg. Covid case missed - setting covid infected person free because model said no covid
