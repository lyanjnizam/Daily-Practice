````python
# if else ladder

age = int(input("age:"))

if age < 12:
  print("Child")
elif age < 18:
  print("Teen")
elif age < 35:
  print("Young Adult")
elif age < 60:
  print("Adult")
else:
  print("Senior")
````

````python
# BMI CALCULATOR

weight = float(input("weight in kg:"))
height = float(input("height in cm:")) * 0.01

bmi = weight / (height * height)

if (bmi < 16):
  print("Severe Thinness")
elif (bmi < 17):
  print("Moderate Thinness")
elif (bmi < 18.5):
  print("Mild Thinness")
elif (bmi < 25):
  print("Normal")
elif (bmi < 30):
  print("Overweight")
elif (bmi < 35):
  print("Obese Class I")
elif (bmi < 40):
  print("Obese Class II")
else:
  print("Obese Class III")
````
