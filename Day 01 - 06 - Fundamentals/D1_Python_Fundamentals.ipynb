## Type Casting
- it means we can force one type of variable into another type

```python
# Type Casting

x = 12
y = "12"
z = 12.56
int_z = int(z)
str_x = str(x)
float_y = float(y)

print(int_z)
print(str_x)
print(float_y)
```

```python
# by taking input

# BMI CALCULATOR

name = input("name:")
age = int(input("age:"))
weight = float(input("weight in kg:"))
height = float(input("height in cm:")) * 0.01

bmi = weight / (height * height)

print(f"{name}! Your BMI is {bmi:.2f}")
```
