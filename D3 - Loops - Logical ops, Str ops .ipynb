````python
# only way to look inside range is, using a for loop

for i in range(0):
    print(i)
for i in range(5): # ending element is not taken, because we start from 0
    print(i) # i takes values from 0 to 4 === 5 elements
print("-" * 100)
for i in range(0, 10, 2): # range(start, end, stepsize)
    print(i)
for i in range(0, 1, 2):
    print(i)
````

#### Time Counter App

````python
# time package
import time

s = 10
for i in range(s):
    time.sleep(1)
    print(f"{i + 1} second/s")
````

````python
# Rocket timers do a Countdown Timer
import time
for i in range(10, 0, -1):
    time.sleep(1)
    print(f"{i}!")
else:
    print('Blast!!')
````

````python
# enumerate - is a method
# that takes a sequence type or object, and gives numbers to each element
# returns a (index, item) pair
# adds a counter/ index to every element in an iterable

for i in enumerate('hello'):
    print(i)

for idx, item in enumerate('hello'):
    print(item)
````

### while loop

- you have to be careful to set up a condition that ends

```python
while <condition>:
    # code
    # control - increment / decrement

```
- what if a condition is always true, how will the loop stop?

````python
# print name 10 times

d = 0
while d < 10: # always true
    print('name')
    d = d + 1

````

````python
# implement a countdown timer using a while loop

import time
t = 10
while t > 0:
    time.sleep(1)
    print(f"{t}!")
    t = t - 1
else:
    print("TIME OUT!")
````

### Loop Control
- break - stop the loop early
- continue - skips to the next iteration
- pass - does nothing, just helps execute instead of error

````python
for i in range(0, 10):
    print(i)
    if i == 5:
        break
````

````python
for i in range(0, 10):
    print(i)
    if i == 5:
        continue
````

````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

````python
for i in range(10):
    if i == 5:
        continue
    print(i)
````

````python
for i in range(10):
    if (i == 3) or (i == 4) or (i == 5):
        continue
    print(i)
````

````python
for i in range(3):
    if (i == 1) and (i == 2):
        print('Hi!')
        continue
    print(i)
````

- Input Validation - to make sure the correct input goes into a program
    - only the correct input goes through
- Exception Handling - targets the error that can come up

- when writing code for a strong program, or a good software, these two methods are very important

````python
# how to use break, continue, input validation, and exception handling

while True:
    num = (input()) # string

    # check for exit
    if num == 'q':
        print('Exit!')
        break

    # validate if input is an integer
    # consider all wrong things users can send
    # alphabets or words or symbols or numbers
    if not num.isdigit():
        print('Invalid Input!')
        continue

    print("Square:", int(num) ** 2)
    print("Cube:", int(num) ** 3)
````

````python
num = '78'
print(num.isdigit()) # string methods
no = 'two'
print(no.isdigit())
````

````python
while True:
    num = (input())

    if num == 'q':
        print('Exit!')
        break

    # Exception Handling - If we did not have input validation

    try:
        print("Square:", int(num) ** 2)
        print("Cube:", int(num) ** 3)
    except ValueError:
        print('Invald Input!')
````
