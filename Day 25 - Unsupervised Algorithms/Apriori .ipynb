```python
import pandas as pd 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transaction = [
    ['Milk', 'Butter', 'Bread'],
    ['Milk', 'Bread'],
    ['Bread', 'Butter', 'Eggs'],
    ['Milk', 'Eggs'],
    ['Bread', 'Butter', 'Juice']
]

te = TransactionEncoder()

te_array = te.fit_transform(transaction)
te_array = te.fit(transaction).transform(transaction)

te_array
```

```python
te.columns_
```

```python
df_transaction = pd.DataFrame(te_array, columns= te.columns_)
df_transaction
```

```python
frequent_itemsets = apriori(df_transaction,
                            min_support= 0.4,
                            use_colnames= True)
frequent_itemsets
```

- Support: How often the rule appears in the dataset.
- Confidence: Probability of buying item B if item A is bought.
- Lift: Strength of the rule over random chance. (>1 means it's a good rule)

```python
rules = association_rules(frequent_itemsets,
                          metric= 'confidence',
                          min_threshold= 0.7
                          )
rules
```

```python
final_table = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
final_table
```

```python
current_cart = {'Bread'}
recommendation = final_table[rules['antecedents'] == current_cart]

recommendation
```
