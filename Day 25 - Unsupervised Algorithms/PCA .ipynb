# PCA (Principal Component Analysis)

- Dimensionality Reduction technique
- means reducing the dimensions of data
- Why do we need that?
    - when we know that more data means the model will learn more?

Curse of Dimensionality
- Increase in dimensions leads to:
    - Sparsity of data:
        - information becomes less dense and there is an increase of empty spaces that just fill the data without adding any information

    - Increases computional cost:
        - because with addition of dimensions, data expands exponentially
        - 

    - Increases the risk of Overfitting:
        - because we are adding more noise when we fit in high dimensional data
        - and the model learns from the noise as well
        - other words, model memorizes training data

```python
import pandas as pd

from sklearn.decomposition import PCA
```

```python
data = {
    'Height': [170, 165, 180, 175, 160, 172, 168, 177, 162, 158],
    'Weight': [65, 59, 75, 68, 55, 70, 62, 74, 58, 54],
    'Age': [30, 25, 35, 28, 22, 32, 27, 33, 24, 21],
    'Gender': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0]
}
df = pd.DataFrame(data)
df
```

```python
x = df[['Height', 'Weight', 'Age']]

pca_model = PCA(n_components= 1)

principal_component = pca_model.fit_transform(x)
principal_component.round(2)
```
