Different types of EDA:
1. `Univariate` - analysis of only `one` variable
2. `Bivariate` - -- of `two` variables
3. `Multivariate` - of `more than two` variables
- does not only mean for visualization

EDA can be:
- `Quantitative` - Descriptive - Summary tables, central tendencies (mean, median, mode), dispersion, variance
- `Graphical` - Visual - Bar charts, histogram, scatter plot, box plot, etc

# Matplotlib
- 

##### Bar Graph
- bivariate
- Multivariate
- we plot categories against numbers or frequency
- 

```python
import matplotlib.pyplot as plt

# Bar chart
categories = [1, 2, 3, 4, 5]
values = [100, 250, 300, 340, 280]

plt.bar(categories, values,
        color= ["#571297E3", "#BB1C51", "#F6750B", "#1FE79A", "#2B60DC"],
        edgecolor= 'black',
        linewidth= 3,
        alpha= 0.5, # Transparency
        hatch= ['X\\', 'x|||||', '+', 'o/', 'O']
        )

plt.title('Bar Chart of categories')
plt.xlabel('Category')
plt.ylabel('Values')

plt.show()
```

#### Line Plot
- Bivariate
- two axis
- two data arrays
- 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([x for x in range(4)])
y = x**2

plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
```

```python
x = emp.Salary.sort_values()
y = emp.Experience

plt.plot(x, y)
```

```python
x = np.linspace(0, 7, 100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y, color= 'black', label= 'Sine')
plt.plot(x, y2, color= "#740EE2", label= 'Cosine')

plt.legend()
plt.grid()

plt.show()
```

##### Histogram
- Univariate analysis
- generally used to check frequency --- Frequency on y-axis by default
- used for number ranges too
- not just categories
- can be continous bars as well

```python
wc = pd.read_csv('wc_2026_fixtures.csv')

plt.hist(wc['team1_fifa_rank'])
```

```python
import numpy as np
np.random.seed(42)
data = np.random.normal(size= 1000) # 1000 data points
plt.hist(data,
         bins= 7,
         edgecolor= 'black',
         linewidth= 1
         )

plt.title('Normal Distribution')
plt.xlabel('data')
plt.show()
```
