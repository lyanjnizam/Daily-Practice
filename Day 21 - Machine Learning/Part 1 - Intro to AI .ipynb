- 
- Artificial(man made) system or machine that can **mimic human intelligence or behaviour**
- 
- **Artificial Intelligence** - Smart Behaviour can be qualified as AI
- AI is a general field in computer science
- 
- we start from **Traditional Computing**:
    - Programming Languages
    - Follows algorithms: series of steps to solve a problem
    - Uses computer hardware like CPUs

- Any well written program can qualify as AI?
    - because Algorithms solve problems as well?
- We still sometimes like to call some traditional programming as === NOT AI
- 
- eg. we say that if we are only using 'IF ELSE programming' ==== it's NOT AI
    - still to some extent intelligence - but Computers itself are dumb
    - we as humans give these instructions

- If we are the ones doing the `thinking`, then it's NOT AI
    - Machine is not doing any of the intelligent things
    - or not showing intelligent behaviour

            - ANALYZE - analysis
            - THINK - Process Information
            - CONCLUDE - drawing conclusions, or making predictions
            - TAKE DECISIONS - decision making

            - PROBLEM SOLVING - 
            - AUTOMATION - 
- 
- Whatever is just IF ELSE ========= NOT AI
- Whatever shows any or most of these intelligent behaviours ===== AI

## AI vs NOT AI


1. AUTO COMPLETE in our Keyboards. AI or NOT AI? WHY?
    - Earlier Keyboards, Rule Based AI was used. We would not call it AI in today's time frame
    - YES. But the recent keyboard softwares are different.
        - looks at the history of use and predicts the next word
        - It adapt over time 

2. BIO METRIC facial recognition. (Store faces and then compare them with the current one) AI or NOT AI? WHY?
    - **NOT AI**.
        - It does analyse. It only maps the face to an existing database
        - if data match YES, if not NO!

3. FACE ID of phones. AI or NOT? WHY?
    - YES.
        - It not only **analyses** new faces all the time, but it also **learns** the owners face over time.
        - when you buy a new phone and set up face, do you feel your phone struggles initially to recognize you? but after a few months of use, it can very quickly recognize your face and recognize if someone else is trying to unlock.
        - 

4. Self Driving Cars. AI or NOT?
    - YES.
        - SDC almost all of them use **Computer Vision**
        - and Move on their own

5. Machines used in industry. AI or NOT?
    - move to left. do welding. move right move the object. do welding
    - NOT AI.
        - we are ones that gave those instructions.
        - we program those machines to do exactly what we want

6. Drones
    - will be fed instructions to fly
    - NOT AI.
        - But if the drone has vision, and makes decisions of movement on its own, based on what it sees
        - If it has sight part includes **Computer vision**, and moves around based on it,
        - then YES
        - examples: example of an AI Drone
            - Farmer drones with AI enabled **weed detection** (computer vision application - AI)
            - they fly over a field of crops
            - detect weed plants among all the others
            - laser burn them

7. Missiles:
    - used to track heat signature - NOT AI
        - only has a heat sensor which detects the location to get the object
        - based on that moving to the target
        - that's not AI
    - but nowadays, missiles use computer vision (AI) along with heat signatures
    - are AI enabled as well

AI has a lot of subfields:
- Artificial Intelligence - mimic human intelligence
- Machine Learning - data/algorithms -> train -> predictions
- Deep Learning - Neural Networks -> predictions
- Generative AI - Large Language Models, Transformers etc.

**Artificial Narrow Intelligence (ANI)**:
- Specialized in one task, like facial recognition or language translation. Current real-world AI falls into this category.
- Designed for a specific task and cannot operate outside its trained scope.
- Examples:
    - ChatGPT (text-based conversation), Siri, Alexa, facial recognition systems.
    - Robotics Use: Industrial robots performing repetitive tasks or autonomous drones with fixed functions.

**Artificial General Intelligence (AGI)**:
- Hypothetical AI that can perform any intellectual task a human can.
- Theoretical at present.
- Would perform any intellectual task a human can.
- Adapting knowledge across domains without retraining.
- Could enable robots to learn new skills in real-time.

**Artificial Super Intelligence (ASI)**:
- Hypothetical AI that surpasses human intelligence.
- In reasoning, creativity, and emotional understanding.
- Would have independent goals, beliefs, and possibly emotions.
