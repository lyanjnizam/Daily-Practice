## LIST

- data structure -> something that stores data
- 
- List is a versatile data structure that stores different items
- Mutable - You can change values of items inside a list
- Ordered - whatever order you put things in, that is maintained
- Index - you can target elements by indexes
- Heterogenous - can store multiple types of data at once

```python
# creating a list
alpha = ['a', 'b', 'c', 1, 2, 3.14, True, None, [1, 2, 3], (2, 3, 4)]

# accessing elements in a list
print(alpha[0])
print(alpha[6])

# can we access a list within a list??
# i want the 2 within the 8th element
print(alpha[8][1])
alpha[8][0] = 'one'
print(alpha)
```

- When creating list - trick to create big lists easily
- one can `write logic inside lists` as well - List comprehension
- Syntax: `[<expression> for <variable> in <sequence> <condition>]`
    - `[x for x in range(20) if x % 2 == 0]`
    - only practice can help with this

```python
# List Comprehension
num = [range(20)]
print(num)
nums = [x for x in range(20)]
print(nums)
no = [2*x for x in range(20)]
print(no)
nos = [x for x in range(20) if x % 2 == 0]
print(nos)
```

```python
# Slicing ---> [start:end:stepsize]
print(nums[::]) # all elements
print(nums[5:10])
print(nums[:5]) # first 5 elements
print(nums[10:]) # 10 to last
print(nums[-3:])
print(nums[-1]) # last element
print(nums[::-1]) # entire list in reverse
```

#### LIST OPERATIONS
- `append()` - add an item at the end of the list
- `pop()` - take an element from the end of the list - or a specified index
- `remove()` - remove by its value - first instance encountered
- `reverse()` - inverts the list

```python
# create a list of favorite fruits
fav_fruits = ['honeydew', 'watermelon', 'banana', 'grape', 'apple']

# how do I add mango to the list
fav_fruits.append('mango')
print(fav_fruits)
print(type(fav_fruits))
print(len(fav_fruits)) # how many elements in a ds
```

```python
fav_fruits.remove('mango')
print(fav_fruits)
```

```python
fav_fruits.append('grape')
print(fav_fruits)
```

```python
fav_fruits.remove('grape')
print(fav_fruits)
```

```python
last_fruit = fav_fruits.pop() # you can save the popped value inside of a variable as well
print(fav_fruits)
print(last_fruit)
```

```python
last = fav_fruits.pop(2)
print(fav_fruits)
print(last)
```

```python
fav_fruits.insert(0, last)
print(fav_fruits)
```

```python
nums.reverse()
print(nums)
```

## TUPLE - 
- similar to list, but immutable
- use them for data that does not change
    - personal details - name, birthdate
    - `username` of a person on a platform
- are heterogenous

```python
# your location
cordinates = (19, 10)

type(cordinates)
```

```python
print(cordinates[:1]) # slicing or accessing works
print(cordinates[1])
```

```python
# information about books
book1 = ('1984', 'George Orwell', 1949, ['available'])
book2 = ('Atomic Habits', 'James Clear', 2012, ['unavailable'])
book3 = ('Think and Grow Rich', 'Napolean Hill', 2014, ['available'])
```

```python
print(book1[3])
# changing a list within a Tuple ---> Shallow Immutability

book1[3][0] = 'unavailable'
print(book1)
```

```python
# pack or unpack tuples
# data comes in small chunks ---> in form of tuples
# everytime someone fills a form, a tuple is sent to the database
person = 'Alya', '21', 'MY'
print(type(person))

name, age, nationality = person # values from person ---> UNPACKED into 3 different variables
print("Name:", name)
print("Age:", age)
print("Nationality:", nationality)
```
