## Functions
- is a reusable block of code - create once and then call again and again
- performs a specific function - one function should only do one thing - should not solve everything at once
- Function are like a machine
    - we give inputs
    - function does some work
    - gives output

- A function name must start with a letter or underscore
- A function name can only contain letters, numbers, and underscores
- Function names are case-sensitive (myFunction and myfunction are different)

````python
def basic_arthematic_operations(num1, num2):
    sum = num1 + num2
    diff = num1 - num2
    product = num1 * num2
    div = num1 / num2

    return sum, diff, product, div

print(basic_arthematic_operations(20, 10))
type(basic_arthematic_operations(20, 10))
````

````python
def greet(name):
    print(f"Hello! Good Morning {name}!")
    print(f"How are you doing, {name}?")
    print(f"You are doing very well, {name}!")
    print(f"Good job! Well done, {name}.")
    print(f"Nice to meet you, {name}.")
````

````python
greet('Friend')
````

````python
print('EXERCISE 1')

def add(num1, num2):
    sum = num1 + num2
    return sum

add(24, 55)
````

````python
print('EXERCISE 2')

def total(list):
    sum = 0
    for x in list:
        sum += x
    return sum

total([2, 3, 4, 1, 0])
````

````python
print('EXERCISE 3: Sum of first n numbers')

def sum_n(num):
    sum = 0
    for i in range(num + 1):
        sum += i
    return sum

sum_n(10000000)
````

````python
# Local vs Global scope

# Global scope
num = 20
print(num)

def temp(num):
    num = 67 # Local scope
    return num #Local scope

# Global scope
print(num)

print(temp(6))
print(temp(1))
````
