````python
def square(number):
    square_num = number * number
    return square_num
````

````python
square(3)
````

````python
def basic_arthematic_operations(num1, num2):
    sum = num1 + num2
    diff = num1 - num2
    product = num1 * num2
    div = num1 / num2

    return sum, diff, product, div

print(basic_arthematic_operations(20, 10))
type(basic_arthematic_operations(20, 10))
````

````python
def greet(name):
    print(f"Hello! Good Morning {name}!")
    print(f"How are you doing, {name}?")
    print(f"You are doing very well, {name}!")
    print(f"Good job! Well done, {name}.")
    print(f"Nice to meet you, {name}.")
````

````python
greet('Friend')
````
