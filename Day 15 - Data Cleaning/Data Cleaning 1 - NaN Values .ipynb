# Data Cleaning

- First you check for inconsistencies in data
    - Missing Values
    - Duplicates
    - Errors
    - Outliers
- Removal or replacement
- 
- Every dataset has its identity:
    - Data Distribution - how is the data spread about - what is the std deviation, variance
    - Central Tendencies
        - where the centre of the data is located
        - mean median mode --- represent the centre of the data
    - Data Transformations: (Feature Scaling is one of the things we do here)
        - [Standardization vs Normalization](https://www.geeksforgeeks.org/machine-learning/normalization-vs-standardization/)
        - Normalization - Adjust values to a common scale
            - min max scaling
            - entire data squished between 0 and 1
            - This scales the range to [0, 1] or sometimes [-1, 1]
        - Standardization - Convert data to a standard form for a common scale
            - mean 0 and std deviation 1
            - 

- Feature Scaling - take a feature/column and transform it entirely

### Handling Missing Values
- Deletion (Removal of these missing values)
    - Listwise:
        - Rows where any data is missing
    - Pairwise:
        - remove specific fields with missing values, keep the rest
- Imputation (filling in missing values)
    - estimate/calculate/predict values that should have been there
    - should not affect identity of dataset too drastically
    - distribution, central tendencies should remain almost intact
    - Types:
        - Mean/median/mode imputation - put these values in missing places
        - Predictive imputation - using ML we predict missing values first
        - Forward/Backward fill - propagate the values - fill in with the previous or the next one

```python
df_listwised = df.dropna()
display(df_listwised)
display(df_listwised.isna().sum())
df_listwised.info()
```

```python
df_pairwised = df.dropna(subset= 'Product')
display(df_pairwised)
display(df_pairwised.isnull().sum())
df_pairwised.info()
```

```python
df2 = df.dropna(subset= 'Product').reset_index()
df2.Quantity = abs(df2.Quantity)
display(df2)
display(df2.info())
display(df2.isna().sum())
```

```python
df2.Quantity = df2.Quantity.fillna(df2.Quantity.mode()[0])
display(df2)
df2.isna().sum()
```
