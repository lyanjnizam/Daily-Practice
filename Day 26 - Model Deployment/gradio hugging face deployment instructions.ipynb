# Deploying a Project on Hugging face spaces

Create an account on Hugging face:
> https://huggingface.co/

### Step 1: Get Your Access Token
Before you touch the terminal, you need your secret password (API Key/Token).

1. Log in to your account on the Hugging Face website.

2. Click your profile picture in the top right corner and select Settings.

3. On the left sidebar, click Access Tokens.

4. Click the New token button.

Tip: If you only want to download private models, a "Read" token is fine. If you want to upload your own models or datasets, create a "Write" token.

5. Copy the generated token to your clipboard.

### Step 2: Login to Hugging Face from terminal

1. Open Terminal and run:

>pip install -U huggingface_hub

- You need the official Hugging Face Python library installed to use their command-line tool.

2. Then Terminal run:

> hf auth login
- If user is already logged in. Use `hf auth login --force` to force re-login.
3. Use arrow key to select paste access token
4. Paste your API key in terminal (you won't see anything happening, but trust me it is there).
5. Hit Enter

- If you use browser:

1. Follow the Link that appears: https://hf.co/oauth/device
2. Enter the code in the interface that appears
3. Click on Authorize in the interface

At this point, you can pass any data into your Hugging face space
- if that data is a gradio_app.py and the model you will use with it, then that app will run on the space
- make sure to keep important packages in a requirements.txt
- During the space setup you will get an option to add dependencies to the requirements txt file as well

### Step 3: Deploy your Gradio app

- any folder you are in, all files of that folder will be uploaded to a hugging face space

1. Change directory and select a folder
2. Create a gradio script file <filename.py> where you set the imports and gradio working function and the gradio interface code. Then open terminal and run:

> gradio deploy

3. Write the name for app space
4. Select the gradio file `<filename.py>`
5. Select space and hardware - cpu-basic (just hit enter without writing anything as well)
6. no secrets, hit n
7. yes for requirements
8. Add any dependencies: 
    - Add the 'model.pkl' file name
    - add any other files from the folder that the script might need
    - hit enter at end
9. Select no for adding github action - 
    - if set yes, you can change the code in the app whenever you push to github as well

Congrats! you have a live app. share it with your friends. or keep the link on your resume. or share on linkedin as a project
