### Exercise: House Price Prediction using Linear Regression
1. Get dataset
2. Break data into input and target features
3. Train test split
4. Bring our model
5. Train model on training set
6. make new predictions
6. compare y_predictions vs y_test to check model performance using:
    - MAE, MSE, RMSE, and R2

```python
import pandas as pd

df = pd.read_csv('House Price Prediction Dataset.csv')

display(df.head())
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score)

x = df[['Bedrooms', 'Bathrooms', 'Floors', 'YearBuilt']]
y = df['Price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)

model = LinearRegression()

model.fit(x_train, y_train)

y_predicted = model.predict(x_test)

mae = mean_absolute_error(y_test, y_predicted)
print(f"MAE = {mae:.2f}")

mse = mean_squared_error(y_test, y_predicted)
print(f"MSE = {mse:.2f}")

rmse = root_mean_squared_error(y_test, y_predicted)
print(f"RMSE = {rmse:.2f}")

r2 = r2_score(y_test, y_predicted)
print(f"R2 = {r2:.2f}")
```

### R2 Score:
- 
- 
Range:
- 1.0 -> Perfect predictions
- 0.9 -> Excellent
- 0.8 -> Very Good
- 0.7 -> Good
- 0 -> Model is no better than predicting the average
- Negative -> Worse than simply predicting average
