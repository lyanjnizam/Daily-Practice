## Exception Handling

When an error occurs, or exception as we call it, Python will normally stop and generate an error message.

```python
try:
    # code
except SomeError:
    # what to do or error msg

- Common examples of errors
1. Dividing by Zero - ZeroDivisionError
2. Converting str to int - TypeError
3. Opening file that does not exist - FileNotFoundError
4. Code that is not valid - SyntaxError

````python
try:
    x = int(input("enter a number"))
    y = 10 / x
    print("Result:", y)
except:
    print("Invalid!")
````

- Handling errors deliberately
- we don't let the program bomb or crash
- Crashing gracefully

````python
try:
    x = int(input("enter a number"))
    y = 10 / x
    print("Result:", y)
except ZeroDivisionError:
    print("Cannot divide with 0!")
except ValueError:
    print("Number only!")
````

````python
try:
    x = int(input("enter a number"))
    y = 10 / x
    print("Result:", y)
except ValueError:
    print("Number only!")
except Exception as e:
    print(e, 'is Wrong!')
````

````python
try:
    x = int(input("enter a number"))
    y = 10 / x
    print("Result:", y)
except ZeroDivisionError:
    print("Cannot divide with 0!")
except Exception as e:
    print(e)
````

````python
try:
    x = input("enter a number")
    y = 10 / x
    print("Result:", y)
except ZeroDivisionError:
    print("Cannot divide with 0!")
except ValueError:
    print("Number only!")
except Exception as e:
    print(e)
````

- Most common and important use of exception handling you will see in:
    - say your program depends on a website and that website is down
        - you keep that line of code in a try except block
    - if your program depends on a file
        - csv files - contain - what if it does not exists
        - the LOC accessing this file should be kept in try except as well

## File Handling:
- how does python access a file
- operations like:
    - READ
    - WRITE
    - OPEN
    - CLOSE

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

````python
# CREATE - a file
file = open("file.txt", "w") # write - "w"
````

````python
# OPEN - a file
try:
    file = open("aiml.ipynb","r") # read - "r"
except Exception as e:
    print(e)
````

````python
try:
    file_5 = open("D5 - Functions .ipynb","r")
except Exception as e:
    print(e)
````

````python
# WRITE - on a file
# open in write mode
file = open("file.txt", "w+") # write + read

file.write("name, age, phone, subject\n")

file.write("alya, 21, 601157522875, AIML\n")
````

````python
file = open("file.txt", "r+") # read + write
````

````python
file.read()
````

Context Manager - using the `with`
- with open( <filename> , "r+") as file:
        file.read()
        file.write()
        file.close() # is done automatically and you don't need to close files

- automatically closes

````python
with open("file.txt", "r+") as f:
    f.write("Hello People!\n")
    f.write("May peace always be upon you\n")
````

````python
with open("file.txt", "a") as f:
    f.write("Hello People!\n")
    f.write("May peace always be upon you\n")
````

````python
with open("file.txt", "r+") as f:
    content = f.read()
    print(content)
````
