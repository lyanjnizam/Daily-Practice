```python
EVALUATE
    TOPN(10, 'Global Superstore')

// DEFINE
//    MEASURE 'table name'[Measure Name] = function('Table Name'[Column name])
DEFINE
    MEASURE 'Global Superstore'[Total Profit2] = SUM('Global Superstore'[Profit])

EVALUATE
    {[Total Profit2]}

DEFINE
    MEASURE 'Global Superstore'[Total Cost2] = SUM('Global Superstore'[Sales]) - SUM('Global Superstore'[Profit]) - SUM('Global Superstore'[Shipping Cost])

EVALUATE
    {[Total Cost2]}

EVALUATE
    SELECTCOLUMNS(
        'Global Superstore',
        "First Name", 'Global Superstore'[Customer First Name],
        "Product", 'Global Superstore'[Product Name],
        "Sales", 'Global Superstore'[Sales]
    )
    ORDER BY [Sales]

EVALUATE
    'Global Superstore'
ORDER BY
    'Global Superstore'[Sales] DESC

EVALUATE
    SUMMARIZE(
        'Global Superstore',
        'Global Superstore'[Category],
        "Total Sales", SUM('Global Superstore'[Sales]),
        "Total Costing", SUM('Global Superstore'[Cost])
    )
    ORDER BY 'Global Superstore'[Total Cost]

EVALUATE
    FILTER(
    SUMMARIZE(
        'Global Superstore',
        'Global Superstore'[Category],
        "Total Sales", SUM('Global Superstore'[Sales]),
        "Total Costing", SUM('Global Superstore'[Cost])
    ),
    [Total Costing] > 3000000
    )
    ORDER BY [Total Costing] DESC

EVALUATE
    FILTER(
        TOPN(50, 'Global Superstore'),
        'Global Superstore'[Cost] > 100
    )
    ORDER BY [Cost]
```
