# Data Visualization
- these are visuals made from data
- Data which is generally in form of tables is:
    - complex to interpret
    - hard to find patterns in
    - difficult to do analysis on it
- Visualization are easy to interpret
    - everyone can see them and draw conclusions
    - people with less tech knowledge should be shown/given visualizations instead of tables
    - includes often times the higher executives of companies
- 
- everybody else would see data tables, but you as an analyst should see opportunity
    - to write a really interesting story
    - DATA STORYTELLING is what makes a great analysts different from an average
- Dashboard - is a page/collection of multiple visualizations and annotations
    - are used to show matrics, KPIs and performances
    - Generally a data analysts thinks of a data storyboard and from there comes up with a dashboard
    

- Learn about visualizations and their code here:
[Visualization Gallery](https://python-graph-gallery.com/)

Different types of EDA:
1. `Univariate` - analysis of only `one` variable
2. `Bivariate` - -- of `two` variables
3. `Multivariate` - of `more than two` variables
- does not only mean for visualization

EDA can be:
- `Quantitative` - Descriptive - Summary tables, central tendencies (mean, median, mode), dispersion, variance
- `Graphical` - Visual - Bar charts, histogram, scatter plot, box plot, etc

##### Bar Graph
- bivariate
- Multivariate
- we plot categories against numbers or frequency
- 

```python
import matplotlib.pyplot as plt

categories = [1, 2, 3, 4, 5]
values = [100, 250, 300, 340, 280]

plt.bar(categories, values,
        color= ["#571297E3", "#BB1C51", "#F6750B", "#1FE79A", "#2B60DC"],
        edgecolor= 'black',
        linewidth= 3,
        alpha= 0.5, # Transparency
        hatch= ['X\\', 'x|||||', '+', 'o/', 'O']
        )
plt.title('Bar Chart of categories')
plt.xlabel('Category')
plt.ylabel('Values')

plt.show()
```

### Boxplot
- univariate
- A box plot, also known as a **box-and-whisker** plot, is a graphical representation
- used in descriptive statistics to display the distribution of a dataset
- based on a five-number summary:
    - minimum,
    - first quartile (Q1),
    - median,
    - third quartile (Q3), and
    - maximum.

## Heatmap
- multivariate analysis
- correlation matrix
- you have two numerical columns, or multiple, you can check the value of correlation between the columns, and get a matrix
- Correlation - 
    - how is one variable varying with respect to another
    - `x` will vary proportionately with `x` if `x` = `x`
    - `x` will vary inversely to `y` if `x` = `1/y`
    - 
    - r = 1 (positive correlation) -> one increase, other increase
    - r = 0
    - r = -1 (negative correlation) -> one increase, other decrease
    - and values between 1 and -1
    - examples:
        - is there is relationship between temperature and sweater sales?
            - lower temprature, higher sales ===> Negative
            - we can't say r = -1 ===> - 0.85 === Strong negative
            - 

```python
corr_matrix = df_iris.drop('species', axis= 1).corr().round(2)
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot= True, cmap= 'magma', linewidth= 0.5, linecolor= 'black', vmin= -1, vmax= 1, cbar_kws= {'shrink':0.8})
plt.show()
```
