## Subplots
- are combination of multiple visualizations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style= 'darkgrid', palette= 'viridis', font= 'monospace', font_scale= 0.5, rc= {'figure.figsize': (12, 6)})

data = pd.DataFrame({'Age': [23, 45, 34, 42, 25, 36, 52, 39, 40, 30],
                     'Salary': [4000, 8000, 6000, 7500, 4200, 6200, 9500, 7100, 8500, 5300]
                     })

sns.scatterplot(data= data,x= 'Age', y= 'Salary', color= 'purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.show()

sns.histplot(data['Age'], bins= 6, kde= True, color= 'blue')
plt.title('Histogram of Age')
plt.show()

sns.boxplot(data['Salary'], color= 'red')
plt.title('Box Plot of Salary')
plt.show()
```

- SUBPLOT

```python
plt.figure(figsize= (12, 4))

plt.subplot(1, 3, 1)
sns.scatterplot(data= data, x= 'Age', y= 'Salary', color= 'purple')
plt.title('Age vs Salary (Scatter Plot)')

plt.subplot(1, 3, 2)
sns.histplot(data['Age'], bins= 6, kde= True, color= 'blue')
plt.title('Histogram of Age')

plt.subplot(1, 3, 3)
sns.boxplot(data['Salary'], color= 'red')
plt.title('Box Plot of Salary')

plt.show()
```

### Using the SUBPLOTS method
- Saves a canvas in a fig, ax tuple
- to build a subplot, simply index the ax[]
- because in subplots, you already defined what indexes we can mean
- 
- plt.subplots(2, 2) ===> 2 rows and 2 columns ===> ax[0, 0], ax[0, 1], ax[1, 0], ax[1, 1]

```python
fig, ax = plt.subplots(1, 3, figsize= (12, 4))

ax[0].scatter(x= data['Age'], y= data['Salary'], color= 'purple')
ax[0].set_title('Age vs Salary (Scatter Plot)')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Salary')

ax[1].hist(data['Age'], bins= 6, color= 'blue')
ax[1].set_title('Histogram of Age')
ax[1].set_xlabel('Age')

ax[2].boxplot(data['Salary'])
ax[2].set_title('Box Plot of Salary')
ax[2].set_ylabel('Salary')

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize= (12, 4))
sns.set_theme(style= 'dark', palette= 'viridis', font= 'monospace')

sns.scatterplot(data= data, x= 'Age', y= 'Salary', ax= ax[0])
ax[0].set_title('Age vs Salary (Scatter Plot)')

sns.histplot(data['Age'], bins= 6, kde= True, ax= ax[1])
ax[1].set_title('Histogram of Age')

sns.boxplot(data['Salary'], ax= ax[2])
ax[2].set_title('Box Plot of Salary')

plt.suptitle('The Dashboard', fontsize= 24, fontweight= 'bold', y= 1.05)
plt.tight_layout() # prevents overlapping
plt.show()
```
