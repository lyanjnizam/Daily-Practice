### EDA (Exploratory Data Analysis) with Pandas

- first step in data analysis and ML too
- three parts:
    1. Data Cleaning - identify and remove missing values, or outliers etc
    2. Data Summarization - check the dataset characteristics and descriptive statistics
    3. Data Visualization - visualize complex data into visuals for easy interpretation

- one of the main job responsibilities of a Data Analyst

- some methods that we can run on dataframes
- Basic Inspection methods: (Data Summarization)
    - `df.head()` - gives first 5 elements
        - an argument of 10 gives first 10 elements
    - df.tail() - last 5 but argument can change it
    - `df.sample()` - one random entry from dataset
        - argument can change that too
        - used to check how datasets look like inside
- Basic attributes of a dataframe:
    - `df.shape`
    - df.size
    - df.ndim
    - `df.dtypes`

- Methods for Statistical information as well
    - `df.info()` - general info about df
        - how many records, columns, data type of columns
        - also gives an idea if there are any missing values in data
    - `df.describe()` - gives statistical measurements of columns like count, mean, min, max, std, percentiles

## Filtering
- filter rows and columns
- uses boolean indexing
- keep only the rows you care about
- 
- df[condition] # or mask

```python
# show matches played in Canada
df[df['country'] == 'Canada']
```

```python
# data for matches in which team1_fifa_rank was <= 10
df[df['team1_fifa_rank'] <= 10]
```

```python
# data for matches in which both teams were < 10 fifa rank
df[(df['team1_fifa_rank']<10) & (df['team2_fifa_rank']<10)]
```

```python
# all matches where Mexico is playing,
# as team1, AND the venue is 'Estadio Azteca'
df[(df['team1'] == 'Mexico') & (df['venue'] == 'Estadio Azteca')]
```

```python
# all matches played in either 'Toronto' OR 'Vancouver'
df[(df['city'] == 'Toronto')|(df['city'] == 'Vancouver')]
```

```python
# matches where:
# team1 ranked strictly better (lower number) than team2 AND
# match is either in Mexico OR Canada
df[(df['team1_fifa_rank'] < df['team2_fifa_rank']) & ((df['country'] == 'Mexico') | (df['country'] == 'Canada'))]
```

```python
mask = (df['team1_fifa_rank'] < df['team2_fifa_rank']) & ((df['country'] == 'Mexico') | (df['country'] == 'Canada'))
mask
```

```python
df2 = df.loc[mask, ['team1', 'team2', 'country']]
df2
```

## Aggregation
- Sometimes we need to get aggregate values of columns
- method: sum(), min(), max(), mean()
- 
- AXIS --- 0 >>>>> Operating column wise
- AXIS --- 1 >>>>> of rows

Use of axis argument 3 places:
- `Aggregate` --- Numpy 2D array --- DataFrame
    - axis = 0 ---> Column wise aggregates
    - axis = 1 ---> Row wise aggregates
- `Drop` a column or row in dataframes
    - axis = 0 ---> # Drop ROW
    - axis = 1 ---> # Drop COLUMN

- `Series.agg`(['mean', 'max', 'min', 'count', 'sum', 'std', 'var', 'prod'])

```python
print(df.team1_fifa_rank.max())
df.team1_fifa_rank.min()
```

```python
print(df[['team1_fifa_rank', 'team2_fifa_rank']].min())
print(df[['team1_fifa_rank', 'team2_fifa_rank']].max())
print(df[['team1_fifa_rank', 'team2_fifa_rank']].mean())
print(df[['team1_fifa_rank', 'team2_fifa_rank']].median())
df[['team1_fifa_rank', 'team2_fifa_rank']].mode().iloc[0]
```

```python
df[['team1_fifa_rank', 'team2_fifa_rank']].mean(axis= 1)
```

```python
df['team2_fifa_rank'].agg(['mean', 'min', 'max'])
```

```python
# drop all rows with missing values
df.dropna(inplace= True) # Original effected
print(df.info())
```

```python
cond = (df['team1_fifa_rank']<=5) | (df['team2_fifa_rank']<=5)
df.loc[cond, ['team1', 'team2', 'team1_fifa_rank', 'team2_fifa_rank']]
```

- 
- using df.group('category').count() - gives all columns for each row
- using df.group('category').size() - gives only categories and count

#### Combine DataFrames
- two dataframe and you want to combine them together
- 

JOINS - 
- are used to combine two tables
- this happens along a KEY column ---> PRIMARY KEY --- we cover this in SQL
- 
- https://tse4.mm.bing.net/th/id/OIP.lRkZ1jamrvFYbSdGu3nF5gAAAA?w=350&h=350&rs=1&pid=ImgDetMain&o=7&rm=3

```python
stu = {
    'StudentID': [101, 102, 103],
    'Name': ['Aman', 'Sara', 'Johan']
}
df_stu = pd.DataFrame(stu)
display(df_stu)

marks = {
    'StudentID': [102, 103, 104],
    'Math': [85, 90, 78],
    'English': [88, 92, 80]
}
df_marks = pd.DataFrame(marks)
display(df_marks)
```

```python
df_merged = pd.merge(df_stu, df_marks, on= 'StudentID', how= 'inner')
df_merged
```

```python
df_lefted = pd.merge(df_stu, df_marks, on= 'StudentID', how= 'left')
df_lefted
```

```python
df_righted = pd.merge(df_stu, df_marks, on= 'StudentID', how= 'right')
df_righted
```

```python
df_full = pd.merge(df_stu, df_marks, on= 'StudentID', how= 'outer')
df_full
```

```python
pd.concat([df_stu, df_marks], ignore_index= True)
```
