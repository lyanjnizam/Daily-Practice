## Tkinter

```python
# Tkinter --- Built-in python - no new installation is needed
import tkinter as tk # aliases

# create a blank window
window = tk.Tk() # class is used to create the main application window

# created some text, to attach with the window
greeting = tk.Label(window, text = "Hello User Interface!")

# packing - take the greeting, and put it on the window
greeting.pack()

# Keeps the window open
window.mainloop()
```

```python
import tkinter as tk

window = tk.Tk()
window.title("My First Tkinter GUI App")
window.geometry("720x360") # sets the window size

greeting = tk.Label(window,
                    text = "Hello User Interface!",
                    font = ("Helvetica", 16, "bold"), # font, fontsize, style
                    fg = "white", # foreground
                    bg = "blue" # background
                    )
greeting.pack()
greeting = tk.Label(window, text = "Hello User Interface!", font = ("Helvetica", 32, "bold"))
greeting.pack()
greeting = tk.Label(window, text = "Hello User Interface!")
greeting.pack()

# action
def button_home():
    print("Welcome Back!")

# Button
action_button = tk.Button(window,
                          text = "Home",
                          command = button_home,
                          font = ("Helvetica", 16, "bold"),
                          fg = "black",
                          bg = "yellow",
                          width = 15,
                          height = 3,
                          relief = "raised",
                          cursor = "hand2"
                          )
action_button.pack(padx = 0)

def button_board():
    print("Back to Work!")

action_button = tk.Button(window,
                          text = "Board",
                          command = button_board,
                          font = ("Helvetica", 16, "bold"),
                          fg = "white",
                          bg = "blue",
                          width = 15,
                          height = 3,
                          relief = "sunken", # flat, raised, sunken, solid, ridge, groove
                          cursor = "hand2" # change the mouse to a hand
                          )
action_button.pack(pady = 20)

window.mainloop()
```

### Let's create a calculator
- 

```python
window = tk.Tk()
window.title("Calculator")
window.geometry("500x400") # width x height

label1 = tk.Label(window, text = "Enter first Number", font = ("Helvetica", 12))
label1.pack()

# First input
input1 = tk.Entry(window)
input1.pack(pady = 10)

label2 = tk.Label(window, text = "Enter second Number", font = ("Helvetica", 12))
label2.pack()

input2 = tk.Entry(window)
input2.pack(pady = 10)

# function
def calculate():
    try:
        num1 = float(input1.get())
        num2 = float(input2.get())

        add = num1 + num2
        sub = num1 - num2
        mul = num1 * num2
        div = num1 / num2

        output_label.config(text = f"Addition: {add}\nSubtraction: {sub}\nMultiplication: {mul}\nDivision: {div}")

    except Exception as e:
        output_label.config(text = f"ERROR: {e}")

# Calculate
calculate_button = tk.Button(window, text = "Calculate",
                             command = calculate,
                             font = ("Helvetica", 16, "bold"),
                             width = 8,
                             height = 2,
                             relief = "raised",
                             cursor = "hand2"
                             )
calculate_button.pack(pady = 20)

# Output Label
output_label = tk.Label(window, text = "RESULTS", font = ("Helvetica", 12), fg = "green")
output_label.pack()

window.mainloop()
```

### API connected GUI project
- Build a Tool that can generate a quote, a joke, or an advice at the click of a button
- click Joke, a joke appears
- click Advice, an advice appears

```python
random_joke_url = "https://official-joke-api.appspot.com/random_joke"

random_advice_url = "https://api.adviceslip.com/advice"
```

```python
import requests
response = requests.get(random_joke_url)
data = response.json()
setup = data['setup']
punchline = data['punchline']
print(f"{setup}\n{punchline} :P")
```

```python
import requests
response = requests.get(random_advice_url)
data = response.json()
print(data['slip']['advice'])
```

```python
def fetch_joke():
    try:
        url = random_joke_url
        response = requests.get(url)
        data = response.json()
        setup = data['setup']
        punchline = data['punchline']
        print(f"{setup}\n{punchline} :P")
    except Exception as e:
        return f"ERROR: {e}"
    
fetch_joke()
```

```python
def fetch_advice():
    try:
        url = random_advice_url
        response = requests.get(url)
        data = response.json()
        print(data['slip']['advice'])
    except Exception as e:
        return f"ERROR: {e}"

fetch_advice()
```

```python
import tkinter as tk
import requests

window = tk.Tk()
window.title("API connected GUI project")
window.geometry("750x350")

def fetch_joke():
    try:
        url = random_joke_url
        response = requests.get(url)
        data = response.json()
        setup = data['setup']
        punchline = data['punchline']
        joke_label.config(text=f"{setup}\n{punchline} :P")
    except Exception as e:
        joke_label.config(text=f"ERROR: {e}")

joke_button = tk.Button(window,
                         text = "Joke",
                         command = fetch_joke,
                         fg = "black",
                         bg = "blue",
                         width = 15,
                         height = 3,
)
joke_button.pack(pady = 20)

def fetch_advice():
    try:
        url = random_advice_url
        response = requests.get(url)
        data = response.json()
        advice_label.config(text=data['slip']['advice'])
    except Exception as e:
        advice_label.config(text=f"ERROR: {e}")

advice_button = tk.Button(window,
                         text = "Advice",
                         command = fetch_advice,
                         fg = "black",
                         bg = "yellow",
                         width = 15,
                         height = 3,
)
advice_button.pack(pady = 20)

joke_label = tk.Label(window, text = " ", font = ("Helvetica", 16), fg = "black")
joke_label.pack(pady = 20)

advice_label = tk.Label(window, text = " ", font = ("Helvetica", 16), fg = "green")
advice_label.pack()

window.mainloop()
```
