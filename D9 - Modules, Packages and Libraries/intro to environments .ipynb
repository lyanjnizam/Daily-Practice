### Environments in Python

- In Python, environments refer to isolated **workspaces** where you can **install** and **manage** packages without affecting the global Python installation


- when we are building software, or projects, which version of python we use, is important
    - say:
        - if i write some LOC in python 3.10
        - the same line may not work in python 3.13
        - the developers could have removed something (deprecated) that LOC
        - or the new update of python might have a new way of doing the same thing
        - this is true for all kinds of packages/libraries
- libraries get updated all the time
    - does that mean
        - we built something in 2020
        - if libraries get updated a lot
        - will the same package work in 2026

- DEPENDENCIES - all the things needed to make an app/package works
    - Every program/application/package that we create will only work if we have all the dependencies installed

Whenever we create a project, we must create a separate **environmen**t for it - 
- an environment will save all the **packages we install**
- all the **dependencies** of the project
- save all the **keys/secrets**
- this environment, will not mess up **global environment** (PC environment)
