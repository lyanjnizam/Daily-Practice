### Array Multiplication

Matrix multiplication is not regular multiplication:
- use the `@` operator like ---- A @ B
- use the `np.matmul(A, B)`

````python
import numpy as np
````

````python
A = np.random.randint(0,10,(2,2))
print(A)
B = np.random.randint(0,10,(2,2))
print(B)

A * B
````

````python
A @ B
````

````python
np.matmul(B, A)
````

### Aggregations
- Aggregations: - data and you summarize it
    - Sum
    - mean
    - min max
    - Standard Deviation
    - Variance
    - count
    - argmax argmin

- Aggregation along an AXIS
- `axis = 0` ===> down the ROWs --- (operate on Columns)
- `axis = 1` ===> across the COLUMNs --- (operate on Rows)

````python
data = np.random.randint(0, 10, (2, 3))
print(data)
````

````python
print(data.sum()) # numpy array methods
print(sum(data)) # python way of data aggregation

print(data.mean())
print(data.std())
print(data.var())

print(data.min())
print(data.max())
````

````python
print(data.sum(axis= 0))
print(data.sum(axis= 1))
````

````python
print(data.mean(axis= 0))
print(data.mean(axis= 1))
````

### Boolean Masking - (Filtering)
- used to filter elements out
- 
- array[mask] ==== mask is a condition
- `array_name[<condition>]`

````python
scores = np.random.randint(0, 101, 6)
scores
````

````python
print(scores > 39)
print(scores[scores > 39])
````

````python
age = np.array([5, 17, 24, 51, 23, 14, 8, 29, 33, 44, 10])
vote_age = age >= 18
age[vote_age]
````

````python
voting_eligibility = np.where(age >= 18, 'Eligible', 'Not-Eligible')
````

````python
list_zip = list(zip(age, voting_eligibility))
````

````python
for age, voting_eligibility in list_zip:
    print(f'{age}: {voting_eligibility}!')
````

````python
age = np.array([5, 17, 24, 51, 23, 14, 8, 29, 33, 44, 10])
voting_eligibility = np.where(age >= 18, 'Eligible', 'Not-Eligible')

combined_array = np.column_stack((age, voting_eligibility))
print(combined_array)
````

7. Automated Grading (where): Use np.where to create a brand new "Report Card" array. If a score in the original array is less than 50, it should say "Fail". If it is 50 or above, it should say "Pass".

````python
# Task 1: The Setup
import numpy as np
np.random.seed(42)

# Task 2: Data Generation
students_scores = np.random.randint(0, 101, (10, 3))
print(students_scores)

# Task 3: Array Inspection
print(students_scores.ndim)
print(students_scores.shape)
print(students_scores.size)

# Task 4: Student IDs
student_IDs = np.arange(100, 110)
print(student_IDs)

# Task 5: aggregate
print(students_scores.mean(axis= 0))

# Task 6: The Audit
print(students_scores[students_scores < 50])

# Task 7: Automated Grading
report_card = np.where(students_scores < 50, 'Fail', 'Pass')
print(report_card)
````
