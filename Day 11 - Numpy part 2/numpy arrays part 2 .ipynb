### Array Multiplication

Matrix multiplication is not regular multiplication:
- use the `@` operator like ---- A @ B
- use the `np.matmul(A, B)`

````python
import numpy as np
````

````python
A = np.random.randint(0,10,(2,2))
print(A)
B = np.random.randint(0,10,(2,2))
print(B)

A * B
````

````python
A @ B
````

````python
np.matmul(B, A)
````

### Aggregations
- Aggregations: - data and you summarize it
    - Sum
    - mean
    - min max
    - Standard Deviation
    - Variance
    - count
    - argmax argmin

- Aggregation along an AXIS
- `axis = 0` ===> down the ROWs --- (operate on Columns)
- `axis = 1` ===> across the COLUMNs --- (operate on Rows)

````python
data = np.random.randint(0, 10, (2, 3))
print(data)
````

````python
print(data.sum()) # numpy array methods
print(sum(data)) # python way of data aggregation

print(data.mean())
print(data.std())
print(data.var())

print(data.min())
print(data.max())
````

````python
print(data.sum(axis= 0))
print(data.sum(axis= 1))
````

````python
print(data.mean(axis= 0))
print(data.mean(axis= 1))
````

### Boolean Masking - (Filtering)
- used to filter elements out
- 
- array[mask] ==== mask is a condition
- `array_name[<condition>]`

````python
scores = np.random.randint(0, 101, 6)
scores
````

````python
print(scores > 39)
print(scores[scores > 39])
````

````python
age = np.array([5, 17, 24, 51, 23, 14, 8, 29, 33, 44, 10])
vote_age = age >= 18
age[vote_age]
````

````python
voting_eligibility = np.where(age >= 18, 'Eligible', 'Not-Eligible')
````

````python
list_zip = list(zip(age, voting_eligibility))
````

````python
for age, voting_eligibility in list_zip:
    print(f'{age}: {voting_eligibility}!')
````

````python
age = np.array([5, 17, 24, 51, 23, 14, 8, 29, 33, 44, 10])
voting_eligibility = np.where(age >= 18, 'Eligible', 'Not-Eligible')

combined_array = np.column_stack((age, voting_eligibility))
print(combined_array)
````

### Nested Loops

```python

for <variable> in <iterable>:
    print(<variables>)

````python
for i in range(5):
    print(i)
````

````python
alist = [3, 5, 7, 9]
name = 'Alya'

for variable in name:
    print(variable)
````

- enumerate object is a sequence
- just like zip

- just like range - range can be type casted into a list
    - others cannot

- we run for loops on them - 

````python
for i, j in enumerate(name):
    print(i, j)
````

````python
for i in range(5):
    for j in range(5):
        print(i)
````

````python
array2d = np.random.randint(0, 10, (2, 4))
array2d
````

````python
for row in array2d:
    for value in row:
        print(value)
````

````python
for row in array2d:
    for value in row:
        print(value, end='  ')
    print()
````

````python
for i in range(len(array2d)):
    for j in range(len(array2d[i])):
        print(f'Row {i}, col {j}: {array2d[i][j]}')
````
