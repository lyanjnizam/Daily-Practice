- Possesions (Class) --- Objects are: Phone, Laptop, Car, House

````python
class Possesion:
    # Class attributes
    num_possesions = 0 # Global variable - shared by all objects of a class

    # Class methods
    # _init_ : is a Constructor --- helps build objects later on
    def __init__(self, kind, age, color, brand):
        # self targets object that will be created later
        # save data for an object
        # Instance attributes - Object attributes
        self.kind = kind
        self.age = age
        self.color = color
        self.brand = brand

        Possesion.num_possesions += 1

    def greet():
        print('Hello!')

    def details(self):
        # this should give most important details of an object
        print(f"Type: {self.kind}\nAge: {self.age}\nBrand: {self.brand}")

    # Getter setter methods!

my_car = Possesion('vehicle', 15, 'Turqoise', 'Perodua')

type(my_car)
````

````python
# create an instance or object
my_phone = Possesion('gadget', 4, 'Orange', 'Mi')
my_pc = Possesion('gadget', 1, 'Black', 'NEC')

# use anything inside of a class we use . operator
Possesion.greet()
Possesion.details(my_pc)
Possesion.num_possesions
````

````python
my_pc.details()
````

### Inheritance and Polymorphism
Inheritance
- a new class (CHILD) can reuse code from an existing class (PARENT)
- 
Polymorphism
- the same method, gives different outputs

#### Inheritance
```python
super()._init_() # is used to call a parent class
```
- example:
- Parent ===== Animal
- Children Classes === Dog, Cat etc

````python
# Parent class
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self._age = age # this tells other developers, to not TOUCH this from outside

    def introduce(self):
        return f"Hi. I am {self.name}! I am a {self._age} years old {self.species}."

    # Getter Method
    def get_age(self):
        return self._age
    
    # Setter Method
    def set_age(self, new_age):
        self._age = new_age
        return f"New age: {new_age}"
    
    def have_birthday(self):
        self._age += 1
        if self._age == 1:
            print(f"It's {self.name}'s {self._age}st birthday!")
        elif self._age == 2:
            print(f"It's {self.name}'s {self._age}nd birthday!")
        elif self._age == 3:
            print(f"It's {self.name}'s {self._age}rd birthday!")
        else:
            print(f"It's {self.name}'s {self._age}th birthday!")
````

````python
# CHILD of ANIMAL
class Dog(Animal):

    def __init__(self, name, age): # this is what we will pass ourselves
        # super() refers to the parent class --- could have used Animal as well
        self.name = name
        self._age = age
        super().__init__(name, 'Dog', age)

    def sound(self):
        print(f"{self.name} says: Woof!")

    # different methods for different classes
    def detect_drugs(self):
        if self._age > 3:
            return "AUWWWWWWWWW WOOF WOOF!"
        else:
            return f"NOT YET! {self.name} is just {self._age}!!"
````

````python
class Cat(Animal):

    def __init__(self, name, age):
        self.name = name
        self._age = age
        super().__init__(name, 'Cat', age)

    def sound(self):
        print(f"{self.name} says: Meowww!")
````

````python
class Parrot(Animal):

    def __init__(self, name, age):
        self.name = name
        self._age = age
        super().__init__(name, 'parrot', age)

    def sound(self):
        print(f"{self.name} says: Hallooo!")
````
