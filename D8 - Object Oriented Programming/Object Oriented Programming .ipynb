### Example - syntax
- Possesions (Class) --- Objects are: Phone, Laptop, Car, House
    - parameters - passes using init - variables that we need to build objects
    - 

````python
class Possesion:
    # Class attributes
    num_possesions = 0 # Global variable - shared by all objects of a class

    # Class methods
    # _init_ : is a Constructor --- helps build objects later on
    def __init__(self, kind, age, color, brand):
        # self targets object that will be created later
        # save data for an object
        # Instance attributes - Object attributes
        self.kind = kind
        self.age = age
        self.color = color
        self.brand = brand

        Possesion.num_possesions += 1

    def greet():
        print('Hello!')

    def details(self):
        # this should give most important details of an object
        print(f"Type: {self.kind}\nAge: {self.age}\nBrand: {self.brand}")

    # Getter setter methods!

my_car = Possesion('vehicle', 15, 'Turqoise', 'Perodua')

type(my_car)
````

````python
# create an instance or object
my_phone = Possesion('gadget', 4, 'Orange', 'Mi')
my_pc = Possesion('gadget', 1, 'Black', 'NEC')

# use anything inside of a class we use - operator
Possesion.greet()
Possesion.details(my_pc)
Possesion.num_possesions
````

## Exercise 2:
- Dog Class
- name. and age for objects
- methods:
    - sound ---> prints woof
    - details ---> print name and age

- Cat Class
- name, and age for objects
- methods:
    - sound ---> prints meow
    - details ---> print name and age

- Animal Class
- name, species, age
- methods:
    - introduce: print name and the species

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def sound(self):
        print("Woof!")

    def details(self):
        print(f"Name: {self.name}\nAge: {self.age}")
````

````python
class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def sound(self):
        print("Meow!")

    def details(self):
        print(f"Name: {self.name}\nAge: {self.age}")
````

````python
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age

    def introduce(self):
        print(f"Name: {self.name}\nSpecies: {self.species}")
````
