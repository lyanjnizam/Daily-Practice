## SET {}
- 
- is an unordered collection of unique items
- very useful when you want to store items that should not repeat

```python
set_A = {1, 2, 3, 4, 4, 3, 2}
print(set_A)

set_B = {3, 4, 8, 5, 9, 6}
print(set_B)
```

```python
# Set Operation
print(set_A | set_B) # Union - is combination of all elements uniquely

print(set_A & set_B) # Intersection - the common elements in the sets
```

```python
# list

names = ["idrees", "afif", "hazira", "idrees", "Luqman", "hazira"]
print(names)

# remove all the duplicates

names = set(names)
print(names)

names = list(names)
print(names)
```

```python
names = ["idrees", "afif", "hazira", "idrees", "Luqman", "hazira"]
print(names)

list(set(names))
```

```python
set_A.add(42) # add an element
print(set_A)

set_A.update(names) # add any iterables
print(set_A)

set_A.remove(42) # remove an element
print(set_A)

set_A.discard(42) # when not sure if element is already removed
# this does not throw an error
print(set_A)

random_element = set_A.pop() # randomly returns an element
print(random_element)
print(set_A)

set_A. clear()
print(set_A)
```

## Dictionary
- {key: value}
- data in key-value pairs
- like real life dictionaries, you look for a word and find its description
- unordered
- 

```python
aliases = {
    "idrees": "Vicious Mango",
    "alya": "Ray Hunter",
    "hariz": "SaySmiley"
}

# 3 tuple list --
aliases = dict([
    ("idrees", "Vicious Mango"),
    ("alya", "Ray Hunter"),
    ("hariz", "SaySmiley")
])

aliases['alya']
```

```python
# session ids

sessions = [("id1", 12312412), 
 ("id2", 37185531),
 ("id3", 239052387)]

sessions = dict(sessions)

sessions['id3']
```

```python
student = {
    "name": "Alya",
    "age": 21, 
    "course": "AIML"
}

student
```

Operations:
- add - adds a new key value pair
- Update - modifying an existing one
- delete - removes a key value pair

```python
student["school"] = 'Nexperts Academy'

student
```

```python
print(student.get('course'))
print(student.get('class')) # this fails safely, does not throw error, if non existing elements
```

```python
student.get('class', 'NO') # is the output if not exists
```

```python
student['course'] = 'DS' # update a value
student
```

```python
student.update({'name': 'Areeqa', 'age': 20}) # update an existing set of values
student
```

```python
del student['name'] # TO REMOVE a key along with its data
student
```

```python
last_session = sessions.popitem() # please remove the last added element, and save in variable
print(last_session)
sessions
```

```python
sessions.clear()
sessions
```

```python
for i in student:
    print(i)

for i in student.items():
    print(i)

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
print(student.items()) # a list of tuples is returned
print(student.keys()) # list of only keys
print(student.values()) # list of values
```
