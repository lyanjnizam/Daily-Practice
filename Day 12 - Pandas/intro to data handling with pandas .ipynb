# Selection using the Location or Labels
##### integer Location --- **.iloc[]**
- `df.iloc[row index]` # access rows
- 
- `df.iloc[row_index_slice]` # [start: end: stepsize]
- INDEXING ----
- `df.iloc[row_slice, column_slice]`

```python
df.iloc[0]
```

```python
df.iloc[0:1]
```

```python
df.iloc[0:5:2]
```

```python
df.iloc[::-1]
```

```python
df.iloc[1:4, 1]
```

```python
df.iloc[1:3, 1:3]
```

```python
# last 2 entries for department and experience
df.iloc[-2:, [1, 3]]
```

##### Label - .loc[]
- focuses on labels
- df.loc[0:3] # targets the name of these rows
- Which are the indexes
- 

```python
df.loc[0]
```

```python
df.loc[:2]
```

```python
df.loc[:2, ['Name','Experience']]
```

### Adding and Modifying Columns
- 

```python
df['email'] = ['alice@mail.com', 'bob@mail.com', 'char_lie@mail.com', 'd_Diana@mail.com', 'eve_01@mail.com']
df
```

```python
df['Department'] = [' Engineering', ' Marketing', ' Analytics', ' HR', ' Marketing']
df
```

```python
df.Salary = df.Salary*10 # Scaling operation
df
```

```python
df.loc['d', 'email'] = 'itsdiana@gmail.com'
df.iloc[1, 1] = 'IT'
df
```
