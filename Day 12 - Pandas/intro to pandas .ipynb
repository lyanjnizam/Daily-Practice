```python
import pandas as pd
```

### Core Objects
- Series - is a one dimensional labelled array (like a single column)
    - series = one column
- DataFrame - a two-dimensional table made of multiple series of same index
    - dataFrame = a sheet of columns and rows

```python
maths = pd.Series([20, 30, 40], dtype= 'int8', name= 'math')
print(maths)
type(maths)
```

```python
eng = pd.Series([20, 30, 40], name= 'english')
py = pd.Series([20, 55, 40], name= 'python')
sc = pd.Series([20, 99, 40], name= 'science')
```

```python
pd.concat([maths, eng, py, sc])
```

```python
df = pd.concat([maths, eng, py, sc], axis= 1)
print(type(df))
df
```

DataFrame - 
- column, rows
- 

```python
# dataframe from dictionary of lists --- each list has to be of same size
data = {
    'students': ['Alya', 'Hanna', 'Areeqa'],
    'Maths': [90, 85, 95],
    'English': [80, 75, 85]
}
df = pd.DataFrame(data)
print(df)
display(df)
```

```python
# dataframe from list of dictionaries --- each dictionary can be of different size
# we can miss elements --- those elements will appear as NaN values
data = [
    {'Name': 'Alya', 'Age': 21, 'City': 'Alor Gajah'},
    {'Name': 'Hanna', 'City': 'Cheras'},
    {'Name': 'Areeqa', 'Age': 20}
]
df = pd.DataFrame(data)
df
```

- NaN values
    - Not a Number
    - aka Missing Values

```python
# dataframe from 2D arrays
data = [
    ['Alya', 21, 'Alor Gajah'],
    ['Hanna', 'Cheras'],
    ['Areeqa', 20]
]
df = pd.DataFrame(data)
df
```

```python
data = [
    ['Alya', 21, 'Alor Gajah'],
    ['Hanna', 22, 'Cheras'],
    ['Areeqa', 20]
]
df = pd.DataFrame(data, columns= ['Name', 'Age', 'City'])
df
```
