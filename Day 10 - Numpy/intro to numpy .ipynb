## Arrays
- Array is a homogenous and linear data structure
    - homogenous - you save `only one type of data` elements inside
    - it has indices - we can use them to access data
    - Contigous blocks of memory - every element comes one after the other
        - Unlike lists, where every element is scattered through the memory
- if we already have lists in python, why do we need arrays?
    - Arrays are very **FAST** compared to list
        - reason is not just how numpy was created, but also the fact that we only store single type of data inside arrays
        - 
- Numpy arrays are better at numerical operation than python lists ---> because of **BROADCASTING**

```python
import numpy as np

num_array = np.array(1)
print(num_array)
num_array
```

```python
nums_array = np.array(range(10))
print(type(nums_array))
print(nums_array)
nums_array
```

```python
# homogenity of arrays

homo_array = np.array(['i', 9, 0.98, 'haho'])
homo_array

# dtype='<U32' ---> this data type says that --- this array won't be as fast
# array lose its special powers of speed and memory optimization
# this is worse than a list
```

```python
nums_list = list(range(1,10))
nums_list
```

```python
nums_list = [n*10 for n in nums_list]
nums_list
```

```python
nums_array*10

# no need for comprehension like list
```

- This is called BROADCASTING
    - one of the special things about Numpy
    - Broadcasting means taking an array and operating it with another
        - following a one to one operation
        - every element operated with another

- numpy lets us do numerical operations really well

List | Array
----|-----
store pointers(addresses of the data) | Store actual data
objects are scattered through the memory | are Contigous (one after the other)
CPU has to go to each object address to access data | CPU accesses base address and then adds e.g. 8 to that address to get the next element
Heterogenous, every object of different type | Homogenous, Multiple types are not allowed
every object has some extra baggage like data type information | dtype of an entire array is same
Broadcasting is not allowed | Broadcasting can be done

- you won't be able to notice too much difference until you have 1000s of elements to deal with
- until the data to deal with is huge, you will not notice how arrays are life savers
