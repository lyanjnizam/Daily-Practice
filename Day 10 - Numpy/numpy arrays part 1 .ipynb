### Array Properties
- what attributes numpy gives you for array object

- ndim - means Dimensions - the number of axis of an array - 
    - arrays can be of N dimensions
    - dimensions are how many axis an array has
        - 1D array only has one axis ---> Row
        - 2D array - 2 axes ---> Rows and Columns
        - 3D array - 3 axes ---> Row, Column, height/depth
            - [[[ x ]]]
- shape - number of rows, number of columns
    - shape shows the number of elements along each axes
    - `(<number of elements in array>,)` -- `(x,)`
    - `(<number of element in Rows>, <number of elements in Cols>)` -- `(x, y)`
    - `(elements along depth, elements along row, elements along cols)` - `(z, x, y)`
- size - number of elements inside
- dtype - the type of array or the elements inside are of which type

```python
import numpy as np
array1d = np.array([1, 2, 3, 4]) # 1D array

array1d.ndim
```

```python
array2d = np.array([[1, 2, 3, 4], [3, 4, 5, 6]])

array2d.ndim
```

```python
array3d = np.array([[[1, 2, 3],[1, 2, 3]],
                    [[1, 2, 3],[1, 2, 3]]])

array3d.ndim
```

```python
array3d.shape
```

##### Broadcasting

```python
import numpy as np

num_4 = np.array([1, 2, 3, 4])
num_3 = np.array([10, 10, 20])

(num_4 * 10) + 2
```

```python
arr_table = np.array([[1, 2],
                      [1, 2],
                      [1, 2],
                      [1, 2],
                      [1, 2]])

print(arr_table.ndim)
print(arr_table.shape)
print(arr_table.size)
print(arr_table.dtype)
```

```python
arr_table * 10
```

```python
arr_table * [10, 100]
```

```python
arr_table * [[10], [100], [1000], [10000], [100000]]
```

### Slicing and Indexing
- [Indexing and slicing Towards data science](https://towardsdatascience.com/mastering-indexing-and-slicing-in-python-443e23457125/)
- [numpy indexing and slicing](https://www.pythonlikeyoumeanit.com/Module3_IntroducingNumpy/BasicIndexing.html)

Slicing - when you want to take a slice out of an array or a list
- array[start : end : step]

Indexing - you target the row or column indexes
- array[row_slice, column_slice]
- we do not just target a slice of row
- we can target a slice of a table
